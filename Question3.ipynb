```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("C:\\Users\\enesb\\Downloads\\DailyActivities.xlsx")
df = data.copy()
```

```python
area_of_interest = list(df['Area of Interest'])
df.drop(['Area of Interest'] , axis = 1 , inplace= True)
```

```python
dfT = df.T
dfT.columns = area_of_interest
dfT
```

```python

cl = list(dfT.columns)
fig , axs = plt.subplots(figsize=(7,10) , nrows=1 , ncols=3)
fig.set_figwidth(14)
fig.set_figheight(7)
count = 0
for ax in axs:
    col_name = list(df.columns)[count]
    ax.pie(df[col_name] , labels = df[col_name] , pctdistance=0.9 , radius=0.8  , textprops={'fontsize': 8.1,'color':"#003366"})
    ax.set_title(col_name)
    ax.legend(cl , loc='best' , prop={'size' : 4}).set_title('How many hours\nspend in\ndaily activities?' , prop={'size' : 5})
    count += 1
```

```python
mylist = []
c = len(dfT.columns)
for i in range(c):
    mylist.append(dfT.values[0][i])
    mylist.append(dfT.values[1][i])
    mylist.append(dfT.values[2][i])
```

```python
total = dfT.sum()
myfig1 = plt.figure()
myfig1.set_figwidth(30)
myfig1.set_figheight(8)

size = 0.3

axs1 = myfig1.add_axes([0.1,0.1,0.9,0.9])
axs2 = myfig1.add_axes([0.1,0.1,0.9,0.9])

axs2_Label = list(dfT.columns)

axs1.pie(mylist , radius=0.9 ,wedgeprops={'width' : 0.3 , 'edgecolor':"w"} , textprops={'size' : 9} , colors=['g','r','b'] , labels=mylist , rotatelabels = True , labeldistance= 1);
axs1.legend(['Charles' , 'Henry' , 'Susan'] , loc = 2)
axs1.set_title('Result')

axs2.pie(total , radius=0.6 , wedgeprops={'width' : 0.3 , 'edgecolor':"w" }  , labels=axs2_Label , labeldistance = 0.6 , textprops={'size' : 7} , rotatelabels=True);
axs2.legend(cl , prop = {'size' : 7} , loc=1).set_title('How many hours\nspend in\ndaily activities?' , prop={'size' : 6})
axs2.set_aspect('equal')
```
